### Addition of periodic signals

#### Definition of periodicity

A real function is said to be periodic if there exists a real number $P > 0$ so that $f(t+P) = f(t)$ for all $t$. The number $P$ is said to be a period of the function.

The sum of two periodic functions is often periodic, but not always. Suppose that $x_{1}(t)$ is periodic with least period $T_{1}$, and $x_{2}(t)$ is periodic with least period $T_{2}$. If $T_{1}$ and $T_{2}$ have a common multiple $T$, then $x_{1}(t) + x_{2}(t)$ is periodic with period $T$. However, $T$ is not necessarily the *least* period of $x_{1}(t) + x_{2}(t)$.

To give an extreme example

$$
\sin(t) + -\sin(t)
$$

has no least period.

#### Criterion for sum of periodic functions

If the periods of two periodic functions do not have a common multiple, then their sum is not periodic. Perhaps the simplest example is 

$$
\sin(t) + \sin(\pi t)
$$

whose terms have least periods $2 \pi$ and $2$ respectively

### Fourier series [\[1\]](https://en.wikipedia.org/wiki/Fourier_series)

#### Definition of Fourier series

Here $f(x)$ denotes a function of the real variable $x$, and $f$ is integrable on an interval $[x_{0}, x_{0} + T]$, for real numbers $x_{0}$ and $T$. Fourier series attempt to represent $f$ in that interval as an infinite sum/series, of harmonically related sinusoidal functions.

Outside the interval, the series is periodic with period $T$ (frequency $1/T$). It follows that if $f$ also has that property, the approximation is valid on the entire real line. We can begin with a finite summation (or partial sum):

$$
f_{N}(x) = \frac{A_{0}}{2} + \sum_{n = 1}^{N} A_{k} \cdot \sin \left(\frac {2\pi kx}{T} + \phi _{k} \right) \quad \text{for integers}\ N\ \geq \ 1
$$

We can also write the function in this equivalent form:

$$
f_{N}(x) = \overbrace{a_{0}}^{A_{0}}/2 + \sum_{k = 1}^N \left( \overbrace{a_{k}}^{A_{k} \sin(\phi_{k})} \cdot \cos \left( \frac{2\pi k x}{T} \right) + \overbrace{b_{k}}^{A_{k} \cos(\phi_{k})} \cdot \sin \left( \frac{2\pi k x}{T} \right) \right) = \sum_{k = -N}^N c_{k} \cdot e^{i \tfrac{2\pi k x}{T}}
$$

where

$$
\frac{2 \pi}{T} = \omega_{0}
$$

So which we could write in te following form too:

$$
f_{N}(x) = \overbrace{a_{0}}^{A_{0}}/2 + \sum_{k = 1}^N \left( \overbrace{a_{k}}^{A_{k} \sin(\phi_{k})} \cdot \cos \left( k \omega_{0} x \right) + \overbrace{b_{k}}^{A_{k} \cos(\phi_k)} \cdot \sin \left( k \omega_{0} x \right) \right) = \sum_{k = -N}^N c_{k} \cdot e^{i k \omega_{0} x}
$$

and where

$$
c_k \ \stackrel{\mathrm{def}}{=} \
\begin{cases}
    \frac{A_{k}}{2i} e^{i \phi_{k}} = \frac{1}{2} (a_{k} - i b_{k}) & \text{for }k > 0 \\
    \frac{1}{2} A_0 = \frac{1}{2} a_{0} & \text{for }k = 0 \\
    c_{|k|}^* & \text{for }k < 0
\end{cases}
$$

The inverse relationships between the coefficients are:

$$
A_{k} = \sqrt{a_{k}^{2} + b_{k}^{2}}, \quad \phi _{k} = \operatorname{arctan} \left( \frac{a_{k}}{b_{k}} \right)
$$

When the coefficients (known as Fourier coefficients) are computed as follows:

$$
a_{k} = \frac{2}{T} \int_{x_{0}}^{x_{0} + T} f(x) \cdot \cos \left( \frac{2\pi k x}{T} \right)\ dx = \frac{2}{T} \int_{x_{0}}^{x_{0} + T} f(x) \cdot \cos \left( k \omega_{0} x \right)\ dx
$$

$$
b_{k} = \frac{2}{T} \int_{x_{0}}^{x_{0} + T} f(x) \cdot \sin \left( \frac{2\pi k x}{T} \right)\ dx = b_{k} = \frac{2}{T} \int_{x_{0}}^{x_{0} + T} f(x) \cdot \sin \left( k \omega_{0} x \right)\ dx
$$

$$
c_{k} = \frac{1}{T} \int_{x_{0}}^{x_{0} + T} f(x) \cdot e^{-i \frac{2\pi k x}{T}}\ dx = \frac{1}{T} \int_{x_{0}}^{x_{0} + T} f(x) \cdot e^{-i k \omega_{0} x}\ dx
$$

### Symmetry properties of Fourier series [\[2\]](https://www.youtube.com/watch?v=YmYU-u_pFZY)

#### 1) Even-Function Symmetry

First we examine the case, where the function is even, so

$$
f(t) = f(-t)
$$

For any even periodic functions, the equations for the Fourier coefficients simplify to the following:

$$
a_{k} = \frac{4}{T} \int_{0}^{T/2} f(t) \cos \left( k \omega_{0} t \right)\ dt
$$

$$
b_{k} = 0 \quad \text{for all }k
$$


#### 2) Odd-Function Symmetry

They call a function *odd*, if it has the following property:

$$
f(t) = - f(t)
$$

In this case, the expressions for the Fourier coefficients are as follows:

$$
a_{k} = 0 \quad \text{for all }k
$$

$$
b_{k} = \frac{4}{T} \int_{0}^{T/2} f(t) \sin \left( k \omega_{0} \right)\ dt
$$


#### 3) Time shifting/Translation

For every real $\tau$, if

$$
f(t') = f(t \pm \tau)
$$

then

$$
c_{k} = e^{\pm i k \omega_{0} \tau} \cdot \frac{1}{T} \int_{t_{0}}^{t_{0} + T} f(t) \cdot e^{-i k \omega_{0} t}\ dt
$$


#### 4) Frequency shifting/Frequency modulation

The frequency shifting property states that if

$$
h(t) = e^{âˆ’i k_{0} \omega_{0} t} \cdot f(t)
$$

then

$$
c_{k} = c_{k - k_{0}}
$$

so

$$
c_{k} = \frac{1}{T} \int_{t_{0}}^{t_{0} + T} f(t) \cdot e^{-i (k-k_{0}) \omega_{0} t}\ dt
$$


#### 5) Time reversal

We examine the case, where

$$
f \left( t' \right) = f(-t)
$$

Then

$$
c_{k} \Longrightarrow c_{-k} = \frac{1}{T} \int_{t_{0}}^{t_{0} + T} f \left( t' \right) \cdot e^{-i (-k) \omega_{0} t'} \ dt'
$$


#### 6) Differentiation

If

$$
h(t) = \frac{df(t)}{dt}
$$

then

$$
c_{k} = i k \omega_{0} \cdot \frac{1}{T} \int_{t_{0}}^{t_{0} + T} f(t) \cdot e^{-i k \omega_{0} t}\ dt
$$


#### 7) Multiplication/Amplitude modulation [\[3\]](https://link.springer.com/content/pdf/bbm%3A978-1-4020-4818-0%2F1.pdf)

If

$$
h(t) = f(t) \cdot g(t)
$$

and the following is true:

$$
c_{k_{f}} = \frac{1}{T} \int_{t_{0}}^{t_{0} + T} f(t) \cdot e^{-i k \omega_{0} t}\ dt,
\quad
c_{k_{g}} = \frac{1}{T} \int_{t_{0}}^{t_{0} + T} g(t) \cdot e^{-i k \omega_{0} t}\ dt
$$

then

$$
c_{k_{h}} = \sum_{i = -\infty}^{\infty} c_{i_{f}} c_{(k-i)_{g}}
$$

### Fourier transform

#### Definition of Fourier transform [\[4\]](https://en.wikipedia.org/wiki/Fourier_transform)

The Fourier transform of the function $f$ is traditionally denoted by adding a circumflex: $\hat{f}$, or using the $\mathcal{F}$ symbol. There are several common conventions for defining the Fourier transform of an integrable function $f:\mathbb{R} \to \mathbb{C}$. Here we will use the following definition:

$$
\hat{f}(\xi) = \int_{-\infty}^{\infty} f(x) e^{-2\pi i x \xi}\ dx
$$

When the independent variable $x$ represents time (noted by $t$), the transform variable $\xi$ represents frequency (noted by $\omega$) E.g. if time is measured in $s$, then the frequency is in $1/s$. Under suitable conditions, $f$ is determined by $\hat{f}$ via the inverse transform:

$$
f(x) = \int_{-\infty}^{\infty} \hat{f}(\xi) e^{2\pi i x \xi}\ d\xi
$$

for any real number $x$.

#### Convolution [\[5\]](http://web.stanford.edu/class/ee102/lectures/samp_mod.pdf)

Suppose $f(x)$, $g(x)$ have the Fourier transform $\hat{f}(\xi)$, $\hat{g}(\xi)$, the *convolution* $y = f * g$ of $f$ and $g$ is given by

$$
y(t) = \int_{-\infty}^{\infty} f(\tau) g(x - \tau)\ d\tau
$$

Therefore

$$
\hat{y}(\xi) = \hat{f}(\xi)\hat{g}(\xi) 
$$

(E.g. convolution in the time domain corresponds to multiplication in the frequency domain.)

### Symmetry properties of Fourier transform [\[6\]](http://fourier.eng.hmc.edu/e101/lectures/handout3/node2.html)

#### 1) Translation/Time shifting

For any real number $x_{0}$, if

$$
h(x) = f(x âˆ’ x_{0})
$$

then

$$
\hat{h}(\xi) = e^{âˆ’2\pi x_{0} \xi} \hat{f}(\xi)
$$

#### 2) Frequency shifting/Frequency modulation

For any real number $\xi_{0}$, if 

$$
h(x) = e^{2 \pi ix \xi_{0}} f(x)
$$

then

$$
\hat{h}(\xi) = \hat{f}(\xi âˆ’ \xi_{0})
$$

#### 3) Time reversal

If

$$
h(x) = f(âˆ’x)
$$

then

$$
\mathcal{F} \left[ {h}(-x) \right] = \hat{h}(-\xi)
$$

#### 4) Conjugation

If $h(x) = \overline{f(x)}$, then

$$
\hat{h}(\xi) = \overline{\hat{f}(-\xi)}
$$

In particular, if $f$ is real, then one has the reality condition

$$
\hat{f}(-\xi) = \overline {\hat{f}(\xi)}
$$

that is, $\hat{f}$ is a Hermitian function. And if $f$ is purely imaginary, then

$$
\hat{f}(-\xi)  = -\overline{\hat {f}(\xi)}
$$

#### 5) Differentiation

Suppose $f(x)$ is a differentiable function, and both $f$ and its derivative $fâ€²$ are integrable. Then the Fourier transform of the derivative is given by

$$
\widehat{f'}(\xi) = 2 \pi i \xi \hat{f}(\xi)
$$

More generally, the Fourier transformation of the $n$th derivative $f^{(n)}$ is given by

$$
\widehat{f^{(n)}}(\xi)= (2 \pi i \xi)^{n} \hat{f}(\xi)
$$

### Test properties of Fourier series numerically<sup>[1](#footnote1)</sup>

Fourier coefficients can be approximated by numeric integration. With the use of the trapezoid formulation in the case of the $N + 1$ 

$$
x_{\nu} = \nu h
\quad
\left( \nu = 0,1, \dots ,N \right),
\quad
h = \frac{2 \pi}{N}
$$

equidistant base point, the following approximate formula is obtained:

$$
a_{k} \approx \tilde{a}_{k} = \frac{2}{N} \sum_{\nu = 1}^{N} f(t_{\nu}) \cos \left( k t_{\nu} \right),
\quad
b_{k} \approx \tilde{b}_{k} = \frac{2}{N} \sum_{\nu = 1}^{N} f(t_{\nu}) \sin \left( k t_{\nu} \right),
\quad
\left( k = 0,1, \dots ,n \right).
$$

By analogy, the following formula can be used for $c_{k}$ coefficients:

$$
c_{k} = \frac{1}{N} \sum_{\nu = 0}^{N - 1} f(t_{\nu}) e^{-i k t_{\nu}} = \sum_{\nu = 0}^{N - 1} f_{\nu} \omega_{N}^{k \nu},
\quad
\left( k = 0,1, \dots ,n \right)
$$

where

$$
f_{\nu} = \frac{1}{N} f(t_{\nu}),
\quad
t_{\nu} = \frac{2 \pi \nu}{N},
\quad
\left( \nu = 0,1, \dots ,N-1 \right),
\quad
\omega_{N} = e^{-\tfrac{2 \pi i}{N}}.
$$

#### Calculate Fourier coefficients

Given a periodic, function $f(t)$ with period $T$, this function returns the
coefficients 

$$
a_{0}, \quad \{a_{1}, a_{2}, \dots \}, \quad \{ b_{1}, b_{2}, \dots \}
$$

such that:

$$
f(t) \approx \frac{a_{0}}{2} + \sum_{k = 1}^{N} \left( a_{k} \cdot \cos \left( \frac{2 \pi k t}{T} \right) + b_{k} \cdot \sin \left( \frac{2 \pi k t}{T} \right) \right)
$$

If Complex is set to True, it returns instead the coefficients

$$
\{c_{0}, c_{1}, \dots \}
$$

such that:

$$
f(t) \approx \sum_{k = -N}^{N} c_{k} \cdot e^{i\tfrac{2 \pi k t}{T}}
$$

where we define

$$
c_{-n} =c_{|n|}^{*}
$$

```python
def FourierCoeff(signal, t, N_max, Complex=False):
    
    '''
    Parameters
    ----------
    signal : the periodic function's value as an np.array
    t: an np.linsapce array of points of time, where the signal got evalueted
    N_max : the function will return the first N_max + 1 Fourier coefficients

    Returns
    -------
    if Complex == False, the function returns:

    a_0 : float
    a, b : np.float arrays describing respectively the cosine and sine coefficients

    =======
    if Complex == True, the function returns:

    c : 1-dimensional complex-valued np.array of size N+1
    '''

    # This function computes the one-dimensional n-point discrete Fourier Transform (DFT) 
    # of a real-valued array by means of an efficient algorithm called the Fast Fourier Transform (FFT)
    y = np.fft.rfft(signal) / t.size

    if(Complex):
        return(y)
    else:
        y *= 2
        return(y[0].real, y[1:-1].real, -y[1:-1].imag)
```

#### Reconstruct original function with Fourier coefficients

```python
def FourierCalcReal(a_0, a, b, t, T, N_max):
    
    # Sampling rate: 1000 Hz/second
    SampRate = t * 1000
    ReconstTime = np.linspace(0, t, SampRate, endpoint=True)
    
    # np.array to store the reconstructed function's values
    ReconstFunc = np.empty(SampRate)
    
    # Calculate f(t) values by f(t) ~= a_0/2 + {Sum}_k=1->N (a_k*cos(2*pi*kt/T) + b_k*sin(2*pi*kt/T))
    for dt in range(0,len(ReconstTime)):
        ReconstFunc[dt] = a_0/2
        
        for k in range(1,N_max):
            ReconstFunc[dt] += a[k-1] * np.cos(2 * np.pi * k * ReconstTime[dt] / T) + b[k-1] * np.sin(2 * np.pi * k * ReconstTime[dt] / T)

    return(ReconstFunc, ReconstTime)
```

#### 1) Square wave with $\tau$ bandwidth and $T$ period

```python
def f_1(t, tau, T):
    '''
    Input parameters:
    
    't': time in seconds
    'tau': bandwidth
    'T': period of signal
    '''
    
    # Sampling rate: 1000 Hz/second
    SampRate = t * 1000
    time = np.linspace(0, t, SampRate, endpoint=True)
    
    # np.array for containting values of current function
    signal = np.empty(SampRate)
    
    i = 0
    while(i < SampRate):
        if(((t/SampRate * i) % T) < tau):
            signal[i] = 1
            i += 1
        else:
            signal[i] = 0
            i += 1

    return(time, signal, T, tau)
```

```python
def f_1_Fourier(tau, T, N_max):
    
    '''
    Input parameters:
    
    'tau': bandwidth
    'T': period of signal
    'N_max': Index of nth harmonic of Fourier series
    '''
    
    # From Shanon theoreom we must use a sampling frequency larger than the maximum
    # frequency you want to catch in the signal
    f_sample = 2 * N_max
    
    # We also need to use an integer sampling frequency, or the
    # points will not be equispaced between 0 and 1. We then add +2 to f_sample
    time = np.linspace(0, T, f_sample + 2, endpoint=False)
    
    # np.array for containting values of current function
    signal = np.empty(f_sample + 2)
    
    i = 0
    while(i < (f_sample + 2)):
        if(((time[f_sample + 1]/(f_sample + 2) * i) % T) < tau):
            signal[i] = 1
            i += 1
        else:
            signal[i] = 0
            i += 1

    return(time, signal, T, tau)
```

#### 2) Sawtooth wave

```python
def f_2(t, T):
    '''
    Input parameters:
    
    't': time in seconds
    'T': period of signal
    '''
    
    # Sampling rate: 1000 Hz/second
    SampRate = t * 1000
    time = np.linspace(0, t, SampRate, endpoint=True)
    
    # np.array for containting values of current function
    signal = np.empty(SampRate)
    
    i = 0
    while(i < SampRate):
        signal[i] = (((t/SampRate) * i) % T) / T
        i += 1
        
    return(time, signal, T)
```

```python
def f_3(t):
    '''
    Input parameter:
    
    't': time in seconds
    '''
    
    T = 2 * np.pi
    
    # Sampling rate: 1000 Hz/second
    SampRate = t * 1000
    time = np.linspace(0, t, SampRate, endpoint=True)
    
    # np.array for containting values of current function
    signal = np.empty(SampRate)
    
    i = 0
    while(i < SampRate):

        if((t/SampRate * i) % T <= T/2):
            signal[i] = 2 * ((t/SampRate * i) % T)
            i += 1
            
        else:
            signal[i] = 2 * (T - ((t/SampRate * i) % T))
            i += 1
        
    return(time, signal, T)
```
