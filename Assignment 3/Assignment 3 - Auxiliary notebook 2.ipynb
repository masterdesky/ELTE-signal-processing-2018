```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
def PositiveNegative():
    
    return -1 if np.random.rand() < 0.5 else 1


def RandomStep(t, SampRate, weight):
    
    RandomStep = np.random.randint(1,10) / (SampRate/t) * PositiveNegative() * weight
    
    return RandomStep

def RandomStepUniform(t, SampRate, weight):
    
    RandomStep = np.random.uniform(-1,1) / (SampRate/t) * PositiveNegative() * weight
    
    return RandomStep

def ModulatorSignalRandom(t, uniform=False, weight=1):
    
    # Sampling rate: 1000 Hz/second
    SampRate = t * 1000
    time = np.linspace(0, t, SampRate, endpoint=True)

    # np.array for containting values of current function
    signal = np.empty(SampRate)

    i = 1
    signal[0] = 0
    while(i < SampRate):

        if uniform:
            signal[i] = signal[i - 1] + RandomStepUniform(t, SampRate, weight)
            i += 1
        else:
            signal[i] = signal[i - 1] + RandomStep(t, SampRate, weight)
            i += 1

    return(time, signal)
```

### Simple Rand-Walk

```python
t_sine = 10000

fig_stand, axes_stand = plt.subplots(figsize=(28,7))

modulator_time, modulator_signal = ModulatorSignalRandom(t_sine, uniform=True)
end_points_stand = modulator_signal[-1]

color = 'red'
axes_stand.plot(modulator_time, modulator_signal, c=color)

plt.show()
```

### Probability-optimized Rand-Walk

```python
MaximumTime = 100

end_points_lin = np.empty(MaximumTime)
end_times_lin = np.linspace(1,MaximumTime, MaximumTime, endpoint=True)

fig_maxend, axes_maxend = plt.subplots(nrows=2, ncols=1, figsize=(28,14))

i = 1
while(i <= MaximumTime):
    
    modulator_time, modulator_signal = ModulatorSignalRandom(i, uniform=False)
    end_points_lin[i-1] = modulator_signal[-1]
    
    color = np.array((np.random.rand(),np.random.rand(),np.random.rand()))
    axes_maxend[0].plot(modulator_time, modulator_signal, c=color)
    
    i += 1

axes_maxend[1].plot(end_times_lin, end_points_lin)

plt.show()
```
