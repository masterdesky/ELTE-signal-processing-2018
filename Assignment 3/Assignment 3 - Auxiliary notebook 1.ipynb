```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='whitegrid')
```

```python
def sincfunc(a_sinc, t_sinc):
    
    SampRate = t_sinc * 2000
    
    sincfunc = np.empty(2 * SampRate)
    timesinc = np.linspace(-t_sinc, t_sinc, 2 * SampRate)

    i = 0
    while(i < 2 * SampRate):

        sincfunc[i] = np.sinc((i - SampRate) * a_sinc * t_sinc / SampRate)
        i += 1
        
    return(sincfunc, timesinc)
```

```python
sincfunc_1, timesinc_1, = sincfunc(a_sinc=1, t_sinc=10)
sincfunc_2, timesinc_2, = sincfunc(a_sinc=2, t_sinc=10)
sincfunc_3, timesinc_3, = sincfunc(a_sinc=np.pi, t_sinc=10)
```

```python
fig_sinc, axes_sinc = plt.subplots(figsize=(14,7))

axes_sinc.plot(timesinc_1, sincfunc_1, label="T = 1 sec", c="red")
#axes_sinc.plot(timesinc_2, sincfunc_2, c="green")
axes_sinc.plot(timesinc_3, sincfunc_3, label="T =  sec", c="blue")

plt.legend()
plt.show()
```
